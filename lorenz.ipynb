```
import numpy as np
import pandas as pd
from scipy.integrate import ode, solve_ivp
```

```
sigma=8
```

```
def sir_ode(times,init,parms):
    beta, rho = parms
    x, y, z = init
    # ODEs
    dx=8*(y-x)
    dy = x*(rho-z)-y
    dz = x*y-beta*z
    return [dx,dy,dz]
```

```
parms = [0.1,0.05]
init = [0.99,0.01,0]
times = np.linspace(0,200,2001)
```

```
sir_sol = solve_ivp(fun=lambda t, y: sir_ode(t, y, parms), t_span=[min(times),max(times)], y0=init, t_eval=times)
```

```
sir_out = pd.DataFrame({"t":sir_sol["t"],"S":sir_sol["y"][0],"I":sir_sol["y"][1],"R":sir_sol["y"][2]})
```

```
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 3.5), subplot_kw={'projection': '3d'})

for ax, sir_ode, c in [(ax1, sir_ode, 'r'), (ax2, sir_ode, 'b'), (ax3, sir_ode, 'g')]:
    ax.plot(sir_ode[:,0], sir_ode[:,1], sir_ode[:,2], c, alpha=0.5)
    ax.set_xlabel('$x$', fontsize=16)
    ax.set_ylabel('$y$', fontsize=16)
    ax.set_zlabel('$z$', fontsize=16)
    ax.set_xticks([-15, 0, 15])
    ax.set_yticks([-20, 0, 20])
    ax.set_zticks([0, 20, 40])
```

```
xline = plt.plot("t","x","",data=sir_out,color="red",linewidth=2)
yline = plt.plot("t","y","",data=sir_out,color="green",linewidth=2)
rline = plt.plot("t","z","",data=sir_out,color="blue",linewidth=2)
plt.xlabel("Time",fontweight="bold")
plt.ylabel("Number",fontweight="bold")
legend = plt.legend(title="Population",loc=5,bbox_to_anchor=(1.25,0.5))
frame = legend.get_frame()
frame.set_facecolor("white")
frame.set_linewidth(0)
```
